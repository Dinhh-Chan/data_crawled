```python
import requests 
import json
from bs4 import BeautifulSoup 
```

```python
url = "https://www.itu.int/hub/?s=&post_type=post"
#news
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
a_list_page = soup.find_all("a", class_="page-numbers" )
max_page = int(a_list_page[-2].text )  
data_crawled =[]
list_link =[]

for page in range(1,max_page+1):
    link = f"https://www.itu.int/hub/page/{str(page)}/?s&post_type=post" 
    respone = requests.get(link)
    soup = BeautifulSoup(respone.text , "html.parser")
    a_all_news = soup.find_all("a", class_= "card-link")
    for a in a_all_news :
        list_link.append(a.get("href"))
    print(list_link)
with open("itu_link_news.json", "w", encoding="utf-8") as f:
    json.dump(list_link, f, ensure_ascii=False, indent=4)
for link in list_link :
    response = requests.get(link)
    print(response.status_code)
    print(link)
    soup = BeautifulSoup(respone.text, "html.parser")
    title = soup.find("h1", class_= "post-title")
    print(title)
    title = title.text.strip() if title else "none"
    published_day = soup.find("span", class_ = "post-info__date")
    published_day = published_day.text.strip() if published_day else "none"
    div_tag = soup.find("div", class_= "post-tags")
    print(div_tag)
    all_a_tag = div_tag.find_all("a")
    list_tag =[]
    for a_tag in all_a_tag :
        list_tag.append(a_tag.text.strip())
    content = soup.find("div", class_ = "wp-editor-wrapper")
    content = content.text if content else "none"
    data = {
        "tag" : list_tag,
        "title": title,
        "link": link, 
        "abstract": "no abstract", 
        "publised_day": published_day,
        "content": content 
    }
    data_crawled.append(data)
    print(data)
with open("itu_news.json", "w", encoding="utf-8") as f:
    json.dump(data_crawled, f, ensure_ascii=False, indent=4)

```
